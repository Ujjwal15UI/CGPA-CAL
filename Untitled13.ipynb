shallow copy verues deep copy

```python
import numpy as np
#copy() fn
#shallow copy    (b=a[0:5])
#deep copy       (b=a.copy())
a= np.array([10,20,30,40,50,60])
#shallow copy
b=a[0:5]
print(a)
print(b)
```

```python
#deep copy
c=a.copy()
c[1]=888
print(a)
print(c)
```

```python
#without seed
print(np.random.randint(1,10,5))
print(np.random.randint(1,10,5))
print(np.random.rand(5))
print(np.random.rand(5))
```

```python
# withseed(same output everyrun)
np.random.seed(42)
print(np.random.randint(1,10,5))
print(np.random.randint(1,10,5))
print(np.random.rand(5))
print(np.random.rand(5))
print(np.random.uniform(1,10,5))
print(np.random.uniform(50,5,5))
```
