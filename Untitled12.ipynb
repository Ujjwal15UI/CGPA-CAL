np.arange()&np.linspace()

```python
import numpy as np
arr = np.arange(0,10)#np.arnage(start,stop,step)
print(arr)
```

```python
#access values
print(arr[0])
print(arr[1])
print(arr[2])
```

```python
#arange vs linespace
#basic Usage
a=np.arange(0,10,2)
b=np.linspace(0,10,5)
print(a)
print(b)
```

```python
#lenght comparison
print(len(a))
print(len(b))
```

#array Filtering

```python
#aray filtering
arr =np.arange(55,65)
print(arr)
arr>58
```

```python
print(arr>58)
```

```python
boolean_arr=arr>58 #creating a mask
arr[boolean_arr]
```

```python
print(arr[boolean_arr])
```

```python
#example array filtering
arr=np.arange(55,65)
print(arr)
mask=arr>58
print(mask)
#multiple cond
print(arr[(arr>57)&(arr <63)])
#marks examples
marks=np.array([67,34,89,32])
passed_boolean=marks>35
print(passed_boolean)
print(marks[passed_boolean])
```
