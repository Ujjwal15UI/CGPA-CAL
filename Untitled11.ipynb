```python
import numpy as np
arr = np.array([1, 2, 3, 4])
lst = [1, 2, 3, 4]
```

```python
#add
print(arr+5)
print([x+5 for x in lst])
```

```python
#multiply
print(arr*2)
print([x*2 for x in lst])
```

```python
#sqr
print(arr ** 2)
print([x ** 2 for x in lst])
```

```python
#mean
print(np.mean(arr))
print(sum(lst)/len(lst))
```

```python
#filter
print(arr[arr % 2 ==0])
print([x for x in lst if x%2==0])
```

```python
#2D
print(np.array([[1,2],[3,4]]))
print([[1,2],[3,4]])
```

```python
#add two
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a+b)
```

```python
a=[1,2,3]
b=[4,5,6]
print([x+y for x,y in zip(a,b)])
```
